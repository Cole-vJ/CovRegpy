```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [15, 12]
sns.set(style='darkgrid')
np.random.seed(0)
```

```python
def normal(mu, sigma, x):
    return (1/np.sqrt(2 * np.pi * sigma ** 2)) * np.exp(-((mu - x)**2)/(sigma ** 2))
```

```python
temp_1 = np.random.normal(1.5, 0.3, 2500)
temp_2 = np.random.normal(1.2, 0.2, 2500)
temp_3 = np.random.normal(1.8, 0.4, 2500)
temp_4 = np.random.normal(1.0, 0.1, 2500)
temp_5 = np.random.normal(2.0, 0.2, 2500)
temp_6 = np.random.normal(2.3, 0.1, 2500)
# plt.hist(temp_1, bins=21, density=True)
# plt.hist(temp_2, bins=21, density=True)
# plt.hist(temp_3, bins=21, density=True)
# plt.hist(temp_4, bins=21, density=True)
# plt.hist(temp_5, bins=21, density=True)
# plt.hist(temp_6, bins=21, density=True)
plt.plot(np.linspace(0, 3, 100), normal(1.5, 0.3, np.linspace(0, 3, 100)), label='High frequency long risk parity', Linewidth=2, zorder=1)
plt.fill(np.append(np.linspace(1.25, 3, 100), np.linspace(3, 1.25, 100)), np.append(normal(1.5, 0.3, np.linspace(1.25, 3, 100)), np.zeros(100)), zorder=1)
plt.plot(np.linspace(0, 3, 100), normal(1.2, 0.2, np.linspace(0, 3, 100)), label='High frequency long\short risk parity', Linewidth=2, zorder=2)
plt.fill(np.append(np.linspace(1.25, 3, 100), np.linspace(3, 1.25, 100)), np.append(normal(1.2, 0.2, np.linspace(1.25, 3, 100)), np.zeros(100)), zorder=2)
plt.plot(np.linspace(0, 3, 100), normal(1.8, 0.4, np.linspace(0, 3, 100)), label='All frequencies long risk parity', Linewidth=2, zorder=3)
plt.fill(np.append(np.linspace(1.25, 3, 100), np.linspace(3, 1.25, 100)), np.append(normal(1.8, 0.4, np.linspace(1.25, 3, 100)), np.zeros(100)), zorder=3)
plt.plot(np.linspace(0, 3, 100), normal(1.0, 0.1, np.linspace(0, 3, 100)), label='All frequencies long\short risk parity', Linewidth=2, zorder=4)
plt.fill(np.append(np.linspace(1.25, 3, 100), np.linspace(3, 1.25, 100)), np.append(normal(1.0, 0.1, np.linspace(1.25, 3, 100)), np.zeros(100)), zorder=4)
plt.plot(np.linspace(0, 3, 100), normal(2.0, 0.2, np.linspace(0, 3, 100)), label='Low frequency long risk parity', Linewidth=2, zorder=5)
plt.fill(np.append(np.linspace(1.25, 3, 100), np.linspace(3, 1.25, 100)), np.append(normal(2.0, 0.2, np.linspace(1.25, 3, 100)), np.zeros(100)), zorder=5)
plt.plot(np.linspace(0, 3, 100), normal(2.3, 0.1, np.linspace(0, 3, 100)), label='Low frequency long\short risk parity', Linewidth=2, zorder=6)
plt.fill(np.append(np.linspace(1.25, 3, 100), np.linspace(3, 1.25, 100)), np.append(normal(2.3, 0.1, np.linspace(1.25, 3, 100)), np.zeros(100)), zorder=6)
plt.plot(1.25 * np.ones(100), np.linspace(0, 4, 100), '--', Linewidth=2)
plt.title('Distribution of Cumulative Returns for Risk Parity Portfolios')
plt.xlabel('Cumulative Returns')
plt.ylabel('Density')
plt.legend(loc='best')
plt.show()
```
